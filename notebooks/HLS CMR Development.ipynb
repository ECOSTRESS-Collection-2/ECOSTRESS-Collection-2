```python
from HLS import HLS2CMRSTAC, HLS2CMR, HLS_CMR_query, S30_CMR_query, S30_CONCEPT, CMR_query, CMR_query_URLs
from datetime import datetime, timedelta, time
import pandas as pd
from dateutil import parser
```

```python
SENTINEL_REPEAT_DAYS = 5
LANDSAT_REPEAT_DAYS = 16
GIVEUP_DAYS = 10
giveup_date = datetime.utcnow().date() - timedelta(days=GIVEUP_DAYS)
giveup_date
```

```python
tile = "11SPS"
start_date = parser.parse("2022-11-01").date()
end_date = parser.parse("2022-12-01").date()
```

```python
CMR = HLS2CMR()
CMR
```

```python
CMR_search_results = CMR.search(
    tile=tile,
    start_UTC=start_date,
    end_UTC=end_date
)

CMR_search_results
```

```python
sentinel_IDs = CMR_search_results[CMR_search_results.sensor == "S30"].groupby("ID").first().reset_index()[["date_UTC", "tile", "ID"]].rename(columns={"ID": "sentinel"})
sentinel_IDs.date_UTC = sentinel_IDs.date_UTC.apply(lambda date_UTC: parser.parse(str(date_UTC)).strftime("%Y-%m-%d"))
sentinel_IDs
```

```python
sentinel_dates = set(sentinel_IDs.date_UTC.apply(lambda date_UTC: parser.parse(str(date_UTC)).date().strftime("%Y-%m-%d")))
sentinel_dates
```

```python
landsat_IDs = CMR_search_results[CMR_search_results.sensor == "L30"].groupby("ID").first().reset_index()[["date_UTC", "tile", "ID"]].rename(columns={"ID": "landsat"})
landsat_IDs.date_UTC = landsat_IDs.date_UTC.apply(lambda date_UTC: parser.parse(str(date_UTC)).strftime("%Y-%m-%d"))
landsat_IDs
```

```python
landsat_dates = set(landsat_IDs.date_UTC.apply(lambda date_UTC: parser.parse(str(date_UTC)).date().strftime("%Y-%m-%d")))
landsat_dates
```

```python
dates = pd.DataFrame({"date_UTC": [start_date + timedelta(days=days) for days in range((end_date - start_date).days + 1)], "tile": tile})
dates.date_UTC = dates.date_UTC.apply(lambda date_UTC: parser.parse(str(date_UTC)).strftime("%Y-%m-%d"))
dates
```

```python
HLS_IDs = pd.merge(landsat_IDs, sentinel_IDs, how="outer")
HLS_IDs
```

```python
listing = pd.merge(dates, HLS_IDs, how="left")
listing
```

```python
listing["sentinel_available"] = listing.apply(lambda row: not pd.isna(row.sentinel), axis=1)
listing
```

```python
date_list = list(listing.date_UTC)
sentinel_dates_expected = set()

for d in date_list:
    if d in sentinel_dates:
        sentinel_dates_expected.add(d)
    
    if (parser.parse(d).date() - timedelta(days=SENTINEL_REPEAT_DAYS)).strftime("%Y-%m-%d") in sentinel_dates_expected:
        sentinel_dates_expected.add(d)

sentinel_dates_expected
```

```python
listing["sentinel_expected"] = listing.apply(lambda row: parser.parse(str(row.date_UTC)).date().strftime("%Y-%m-%d") in sentinel_dates_expected, axis=1)
listing
```

```python
listing["sentinel_missing"] = listing.apply(
            lambda row: not row.sentinel_available and row.sentinel_expected and parser.parse(str(row.date_UTC)) >= parser.parse(str(giveup_date)),
            axis=1
        )

listing["sentinel"] = listing.apply(lambda row: "missing" if row.sentinel_missing else row.sentinel, axis=1)
listing
```

```python
listing["landsat_available"] = listing.apply(lambda row: not pd.isna(row.landsat), axis=1)
listing
```

```python
date_list = list(listing.date_UTC)
landsat_dates_expected = set()

for d in date_list:
    if d in landsat_dates:
        landsat_dates_expected.add(d)
    
    if (parser.parse(d).date() - timedelta(days=LANDSAT_REPEAT_DAYS)).strftime("%Y-%m-%d") in landsat_dates_expected:
        landsat_dates_expected.add(d)

landsat_dates_expected
```

```python
listing["landsat_expected"] = listing.apply(lambda row: parser.parse(str(row.date_UTC)).date().strftime("%Y-%m-%d") in landsat_dates_expected, axis=1)
listing
```

```python
listing["landsat_missing"] = listing.apply(
            lambda row: not row.landsat_available and row.landsat_expected and parser.parse(str(row.date_UTC)) >= parser.parse(str(giveup_date)),
            axis=1
        )

listing["landsat"] = listing.apply(lambda row: "missing" if row.landsat_missing else row.landsat, axis=1)
listing
```

```python
listing = listing[["date_UTC", "tile", "sentinel", "landsat"]]
listing
```
